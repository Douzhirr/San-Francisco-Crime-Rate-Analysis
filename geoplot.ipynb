```python
import geoplot as gplt
import geopandas as gpd
import geoplot.crs as gcrs
import pandas as pd
import pathlib
import matplotlib.pyplot as plt
import mapclassify as mc
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\Zihan0932\Downloads\Police_Department_Incident_Reports__2018_to_Present.csv")
```

```python
#Draw the heat map representing the density of crimes in San Francisco
gdf1 = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))
sf = gpd.read_file(gplt.datasets.get_path("san_francisco"))
count = gdf1
ax = gplt.polyplot(sf, projection=gcrs.AlbersEqualArea(),figsize=(10,10))
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
gplt.kdeplot(count, cmap="Reds", shade=True, clip=sf, ax=ax, cbar=True)
```
